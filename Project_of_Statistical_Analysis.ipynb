**The Mean**

The mean is known as the average number within a dataset. It is calculated by *the sum of the numbers* (*x*) occuring in the dataset, and dividing it by the population size which is denoted as *N*. The following figure is the formula denoting the mean in statistics:The Mean.png

```python
#Calculate the mean of the number of students per class enrolled in a statistis course offered by a university.

number_of_students = [25, 27, 18, 19, 15]
mean = sum(number_of_students)/len(number_of_students)
print(mean)
```

```python
#Calculate the mean of the number of students per class enrolled in a statistis course offered by a university with the uses of a function.

def mean(numbers):
  total = sum(numbers)
  size = len(numbers)
  mean = sum(numbers)/len(numbers)
  return mean

mean(number_of_students)
```

When it comes to the mean, having an extreme outlier in our dataset can have a heavy effect, the following code will give a demonstration of how an outlier can have a big impact of the average number of a dataset.

```python
number_of_students.append(100) #This will be adding a number of 100 students in our previous dataset.
mean(number_of_students)
```

```python
#Calculate the mean of the time it takes for a customer to check out a Starbucks.

number_of_customers = [6, 4, 2, 5, 4, 1, 5]
mean(number_of_customers)
```

```python
#Calculate the mean concerning two model of cars that has gone through six trials of testing, and tell which model averages more.

model_A = [22, 24, 30, 19, 20, 27]
model_B = [21, 25, 27, 23, 21, 28]

if mean(model_A) > mean(model_B):
  print("The average of model A is: ", mean(model_A), "per gallon")
  print("The average of model B is: ", mean(model_B))
  print("Car Model A averages more miles per gallon than Model B")
else:
  print("The average of model A is: ", mean(model_A))
  print("The average of model B is: ", mean(model_B))
  print("Car Model B averages more miles per gallon than Car Model A")
```

**The Median**

This is the value that is within the middle of our data set when we organize the values within our dataset from least to greatest. In our formula:

*i* represents the index point for the median.

Because 50% of our data is less than the middle value, *P = 50*.

*n* is the number of data points.

Because we want find the number right in the middle, we will divide 50 by 100.
The image gives us the formula that is use to calculate the median or the number that is exact right at the middle within our dataset.

The Median.png

The index point gives us the position of the median. If we get a n.5, then we would round it up to the nearest integer.

```python
r_list = [85, 96, 75, 82, 93, 88, 80, 90, 71]
s_list = sorted(r_list)

print("This is the sorted list: ", s_list)

def index_location(n, list):
  p = 50
  index = ((n)*p)/100
  print("The index so for is: ", index)
  if type(index) is float:
    index = index + 0.5
    index = int(index)
    print("The number at the mid-range is: ", list[index-1])
  else:
    print("The index is not a float")

index_location(len(s_list), s_list)
```

**The Midrange**

The midrange is a way to get the number that is at the center of the data without the need of sorting all the numbers like we did for the median, or getting the sum of all the values like we did for the mean.

midrange.png

```python
r_list = [23, 29, 30, 10, 5]

def mid_range(x, y):
  mid_range = (x + y)/2
  return mid_range

print(mid_range(max(r_list), min(r_list)))
```
