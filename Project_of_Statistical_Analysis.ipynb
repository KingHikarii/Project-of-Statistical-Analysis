**Frequency Distributions**

A frequency distribution is a two-column table. In the left-column, it will list each value in the data set from least to greatest. On the right column it will count the number of each value appearing and record the total on the right column.

```python
daily_hammers = [2, 3, 3, 2, 1, 0, 2, 2, 0, 2, 3, 4, 2, 4, 4, 3, 1, 0, 2, 0]
```

```python
from collections import Counter
from tabulate import tabulate

def frequency_distribution(numbers):
    # Count the occurrences of each number using Counter
    freq_dist = Counter(numbers)

    # Sort the numbers from least to greatest
    sorted_freq_dist = sorted(freq_dist.items())

    # Prepare the data for the table
    table_data = [(num, count) for num, count in sorted_freq_dist]

    # Display the table
    print(tabulate(table_data, headers=["Number", "Frequency"], tablefmt="grid"))

    # Calculate the total number of occurrences
    total_occurrences = sum(freq_dist.values())

    return total_occurrences

# Example usage
numbers = daily_hammers = [2, 3, 3, 2, 1, 0, 2, 2, 0, 2, 3, 4, 2, 4, 4, 3, 1, 0, 2, 0]
total = frequency_distribution(numbers)

print(f"\nTotal number of occurrences: {total}")
```

**Histograms**

A histogram is a bar graph of the frequency distribution. In the bar graph, there will be gaps, and these gaps exists because the data is discrete. Histograms can take both discrete and continous numbers. When we are dealing with continuous numbers, there will be no gaps because we are no longer dealing with categories.

*Discrete data* is data that can take numbers that are countable such as 1, 10, 50 rather than 23.54, 99.41241 which are continuous numbers.

```python
print("Develop a histogram for the grades dataset.")

grades_dictionary = {'A': 2, 'B': 12, 'C': 6, 'D': 2, 'F': 1}

import matplotlib.pyplot as plt

# Given dictionary
grades_dictionary = {'A': 9, 'B': 12, 'C': 6, 'D': 2, 'F': 1}

# Extract keys and values
grades = list(grades_dictionary.keys())
frequencies = list(grades_dictionary.values())

# Create a bar chart (histogram) using the data
plt.bar(grades, frequencies, color='skyblue')

# Add labels and title
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.title('Histogram of Grades')

# Show the plot
plt.show()
```

In statistics, if there is ever the case where you hsbr two data values, then for each label, than you should use a grouped column bar chart.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data
data = {'Subject': ['Math', 'Math', 'English', 'English', 'Science', 'Science'],
        'Gender': ['Boy', 'Girl', 'Boy', 'Girl', 'Boy', 'Girl'],
        'Score': [10, 15, 12, 18, 8, 13]}
df = pd.DataFrame(data)

# Pivot the data to group by Team and Round
df_pivot = df.pivot(index='Subject', columns='Gender', values='Score')

# Set the positions of the bars on the x-axis
x = np.arange(len(df_pivot.index))
width = 0.35  # Width of the bars

# Create the grouped bar chart
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df_pivot['Boy'], width, label='Boy', color = 'darkblue')
rects2 = ax.bar(x + width/2, df_pivot['Girl'], width, label='Girl', color = 'pink')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by subject and gender')
ax.set_xticks(x)
ax.set_xticklabels(df_pivot.index)
ax.legend()

# Function to add labels above the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```

Let's say you wanted to get the total of different items. For example, how much lemonade you've sold + total number of apple juice you've sold, then you should used stacked column bar chart to get this, and the height of the bar will be the total of the two stacked items sold together.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Day 1', 'Day 2']
subgroup1_values = [20, 35]
subgroup2_values = [15, 25]

# Create the stacked bar chart
fig, ax = plt.subplots()

# Plot the first subgroup
ax.bar(groups, subgroup1_values, label='Lemonde', color = 'gold')

# Plot the second subgroup, stacked on top of the first
ax.bar(groups, subgroup2_values, bottom=subgroup1_values, label='Apple Juice', color = 'darkred')

# Add labels and title
ax.set_xlabel('Drinks')
ax.set_ylabel('Items Sold')
ax.set_title('Drinks Sold')
ax.legend()

# Show the plot
plt.show()

# Add legend
ax.legend()

# Show the plot
plt.show()
```

**The Median**

*i* represents the index point for the median.

Because 50% of our data is less than the middle value, *P = 50*.

*n* is the number of data points.

Because we want find the number right in the middle, we will divide 50 by 100.
The image gives us the formula that is use to calculate the median or the number that is exact right at the middle within our dataset.

**The Midrange**

The midrange is a way to get the number that is at the center of the data without the need of sorting all the numbers like we did for the median, or getting the sum of all the values like we did for the mean.

midrange.png

```python
r_list = [23, 29, 30, 10, 5]

def mid_range(x, y):
  mid_range = (x + y)/2
  return mid_range

print(mid_range(max(r_list), min(r_list)))
```

**The Mode**

The mode is known as the number that occurs the most frequently within our data.

```python
numbers = [1,1, 2, 2, 2, 10, 10, 100, 100, 100]

from collections import Counter

def mode(numbers):
    print("Current Numbers:", numbers)

    # Count occurrences of each number in the list
    count = Counter(numbers)

    print("Number counts:", count)

    # Find the maximum occurrence count
    max_count = max(count.values())

    # Collect all numbers that have the maximum count
    modes = [num for num, freq in count.items() if freq == max_count]

    return modes

numbers = [1, 1, 2, 2, 2, 10, 10, 100, 100, 100]
print("The Mode(s) is/are:", mode(numbers))
```
